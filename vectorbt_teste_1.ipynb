```python
!pip install yfinance
import yfinance as yf
import vectorbt as vbt
```

```python
#dados = vbt.BinanceData.download("BTCBRL", start="2022-02-10", interval="1h").get()
dados = vbt.YFData.download("BBAS3.SA", start="2021-01-01", interval="1d").get()
```

```python
dados
```

```python
prices = dados["Close"]
media_rapida = vbt.MA.run(prices, 21)
media_lenta = vbt.MA.run(prices, 60)
entradas = media_rapida.ma_above(media_lenta)
saidas = media_rapida.ma_below(media_lenta)
pf = vbt.Portfolio.from_signals(prices, entradas, saidas)
pf.plot().show()
```
